## Compare distributions of UMD and Chat datasets

The following variables are perhaps too different to be used:
* Ethnicity Black
* Ethnicity Hispanic
* Gerd
* tonsilsize 3-4
* age

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
matplotlib.rc('font', family='serif')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load UMD data
umd_file = '../../../data/processed/umd_data_standard_all.csv'
data_umd = pd.read_csv(umd_file)

# Load chat data
chat_file = '../../../data/processed/chat_data_standard_all.csv'
data_chat = pd.read_csv(chat_file)

# Define variables
admission_variables = ['gender_Male', 'ethnicity_Hispanic','ethnicity_Black', 'ethnicity_White', 'ethnicity_Asian', \
                       'term', 'bmi', 'age', 'allergies_Yes', 'asthma_Yes', 'gerd_Yes', 'tonsilsize_3-4', 'zscore']
laboratory_variables = ['nrem', 'rem', 'sleepefficiency', 'arousali', 'tst50co2', 'oahi', 'cai', \
                        'oai', 'ohi', 'lowsao2', 'peakc02', 'tb90']
test_variable = ['ahi']
continuous_variables = ["bmi", "age", 'zscore']
reference_variables = ['reference: osa18', 'reference: psq', 'reference: ess']

# Set AHI threshold
ahi_thresh = 5

# Isolate admission variables
data_umd = data_umd[admission_variables+['ahi']].copy()
data_umd['test'] = pd.Series(data_umd['ahi']>ahi_thresh, index=data_umd.index)
data_chat = data_chat[admission_variables+['ahi']].copy()
data_chat['test'] = pd.Series(data_chat['ahi']>ahi_thresh, index=data_chat.index)
```

```python
# ahi
f,(ax1, ax2) = plt.subplots(1, 2, sharey=True)
sns.countplot(x='test',data=data_umd, ax=ax1)
sns.countplot(x='test', data=data_chat, ax=ax2)
```
